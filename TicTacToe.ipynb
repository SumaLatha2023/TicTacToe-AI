```python
def analyse(board):
    c = [[0,1,2], [3,4,5], [6,7,8], [0,3,6], [1,4,7], [2,5,8], [0,4,8], [2,4,6]]
    for i in range(8):
        if (board[c[i][0]] != 0 and
            board[c[i][0]] == board[c[i][1]] and
            board[c[i][1]] == board[c[i][2]]):
            return board[c[i][0]]
    return 0

def stateBoard(board):
    print("The current state of the board : \n")
    for i in range(9):
        if i != 0 and i % 3 == 0:
            print()
        if board[i] == 0:
            print("_", end=" ")
        elif board[i] == 1:
            print("O", end=" ")
        elif board[i] == -1:
            print("X", end=" ")
    print("\n")

def user1Turn(board):
    while True:
        pos = int(input("Enter 'X' position [1-9] : "))
        if pos < 1 or pos > 9 or board[pos-1] != 0:
            print("Wrong Move. Try again.")
        else:
            board[pos-1] = -1
            break

def user2Turn(board):
    while True:
        pos = int(input("Enter 'O' position [1-9] : "))
        if pos < 1 or pos > 9 or board[pos-1] != 0:
            print("Wrong Move. Try again.")
        else:
            board[pos-1] = 1
            break

def minmax(board,player):
  x = analyse(board)
  if(x != 0):
    return x*player
  pos = -1
  value = -10
  for i in range(9):
    if(board[i] == 0):
      board[i] = player
      score = -minmax(board,player)
      board[i] = 0
      if(score > value):
        value = score
        pos = i
  if(pos == -1):
    return 0
  return value

def compTurn(board):
  pos = -1
  value = -10
  for i in range(9):
    if(board[i] == 0):
      board[i] = 1
      score = -minmax(board,-1)
      board[i] = 0
      if(score > value):
        value = score
        pos = i
  board[pos] = 1

def main():
  choice = int(input("Enter 1 for Single-Player or 2 for Multi-Player : "))
  board = [0, 0, 0, 0, 0, 0, 0, 0, 0]
  if choice == 1:
    print("Computer Mode")
    print("You 'X' vs Computer 'O'" )
    player = int(input("Enter '1' to play first or '2' to play second : "))
    for i in range(9):
      if(analyse(board) != 0):
        break
      if((i+player)%2 == 0):
        compTurn(board)
      else:
        stateBoard(board)
        user1Turn(board)
    if(analyse(board) == 0):
      print("Draw")
    elif(analyse(board) == -1):
      print("You have Won")
    else:
      print("Computer has Won")


  else:
    print("Custom Mode")
    for i in range(9):
      stateBoard(board)
      if(analyse(board) != 0):
        break
      if(i%2 == 0):
        user1Turn(board)
      else:
        user2Turn(board)
    if(analyse(board) == 0):
      print("Draw")
    elif(analyse(board) == -1):
      print("Player 1 has Won")
    else:
      print("Player 2 has Won")

main()
```
